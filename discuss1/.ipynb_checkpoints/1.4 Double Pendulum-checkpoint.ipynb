````julia
using DifferentialEquations
using Plots
using LinearAlgebra
using Flux: gradient
````

$$
\mathcal{H} = \frac{1}{2ml^2}\frac{p_1^2+2p_2^2-2p_1p_2\cos(\phi_1-\phi_2)}{1+\sin^2(\phi_1-\phi_2)} + mgl[4-2\cos(\phi_1) - \cos(\phi_2)]
$$

````julia
const m = 1.0
const l = 1.0
const g = 9.8

# hamiltonian
H(p,q) = 1.0 / (2m*l^2) * (norm(p)^2 + p[2]^2 - 2prod(p)*cos(q[1]-q[2]))/(1+sin(q[1]-q[2])^2) +
        m*g*l*(4-2cos(q[1])-cos(q[2]))

# ∇H
grad!(dx,args...) = dx .= gradient(args...)[1]

pdot(dp,p,q,params,t) = grad!(dp, q -> -H(p,q), q)
qdot(dq,p,q,params,t) = grad!(dq, p ->  H(p,q), p);
````

````julia
# nilai awal
p0 = [3.0,0.0]
q0 = [π/2,π/2]
ts = (0.0,50.0);

# selesaikan
prob = DynamicalODEProblem(pdot,qdot,p0,q0,ts)
@time sol = solve(prob,reltol=1e-9,abstol=1e-10,dt=0.01);
````

````julia
# polar to cartesian coordinate
function polar_to_cart(sol;dt=0.05,vars=(3,4))
    t = 0:dt:sol.t[end]

    p1 = l*map(x->x[vars[1]], sol.(t))
    p2 = l*map(y->y[vars[2]], sol.(t))

    x1 = l*sin.(p1)
    y1 = -l*cos.(p1)
    x2 = x1 + l*sin.(p2)
    y2 = y1 - l*cos.(p2)
    
    return (t, (x1,y1),(x2, y2))
end

ts, cart1,cart2 = polar_to_cart(sol)

plot(cart1...)
plot!(cart2...)
````

````julia
# variation of energy
ΔE = (H(q0,p0) .- map(u -> H(u[1,:],u[2,:]),sol.u)) / abs(H(q0,p0)) #E0 - Ei

plot(sol.t,ΔE)
````

````julia
```julia
using DifferentialEquations
using Plots
using Flux: gradient

const m1 = 1.0
const m2 = 2.0
const l1 = 1.0
const l2 = 2.0
const g  = 9.81

function H(p,q)
  T_a = m2*l2^2*p[1]^2 + (m1+m2)*l1^2*p[2]^2 - 2m2*l1*l2*cos(q[1]-q[2])*p[1]*p[2]
  T_b = 2m2*l1^2*l2^2*(m1+m2-m2*cos(q[1]-q[2])^2)
  V = - (m1+m2)*g*l1*cos(q[1]) - m2*g*l2*cos(q[2])

  T_a/T_b + V
end

grad!(du,args...) = du .= gradient(args...)[1]

pdot(dp,p,q,params,t) = grad!(dp, q -> -H(p,q), q)
qdot(dq,p,q,params,t) = grad!(dq, p ->  H(p,q), p)

p0 = [0.0,0.0]
q0 = [π/3,3π/5]
ts = (0.0,50.0)

prob = DynamicalODEProblem(pdot,qdot,p0,q0,ts)
sol  = solve(prob,Vern7(),reltol=1e-9,abstol=1e-10,dt=0.05)

# polar to cartesian coordinate
function polar_to_cart(sol;dt=0.01,vars=(3,4))
    t = 0:dt:sol.t[end]

    p1 = map(x->x[vars[1]], sol.(t))
    p2 = map(y->y[vars[2]], sol.(t))

    x1 = l1*sin.(p1)
    y1 = -l1*cos.(p1)
    x2 = x1 + l2*sin.(p2)
    y2 = y1 - l2*cos.(p2)

    return (t, (x1,y1),(x2, y2))
end

ts, cart1,cart2 = polar_to_cart(sol)

plot(cart1...)
plot!(cart2...)

# variation of energy
ΔE = (H(p0,q0) .- map(u -> H(u[:][1:2],u[:][3:4]),sol.u)) ./ abs(H(p0,q0)) #E0 - Ei

plot(sol.t,ΔE)
````
